```python
# Challenge 2
# Crear una tarjeta personal. El usuario (mediante input) puede elegir:
# Crear datos nuevos
# Modificar datos
# Eliminar datos

tarjeta_presentacion = {}

while True:
    eleccion = int(input("Que desea hacer:\n1. Crear datos nuevos\n2. Modificar datos\n3. Eliminar datos\n 4. Salir\nOpcion: "))

    if eleccion == 1:
        clave = input("Que datos desea cargar: ")
        valor = input(f"Ingrese el valor de {clave}:")

        tarjeta_presentacion[clave] = valor
        print("\nValor agregado correctamente")
        print(tarjeta_presentacion)

    elif eleccion == 2:
        clave = input("Que dato desea modificar: ")
        valor = input(f"Ingrese el nuevo valor de {clave}:")
    
        tarjeta_presentacion[clave] = valor
        print("\nValor modificado correctamente")
        print(tarjeta_presentacion)

    elif eleccion == 3:
        clave = input("Que dato desea eliminar: ")
        del tarjeta_presentacion[clave]

        print("\nValor eliminado correctamente")
        print(tarjeta_presentacion)

    elif eleccion == 4:
        print("\nSe termino la tarjeta de presentacion")
        print(tarjeta_presentacion)
        break

    else:
        print("\nNo se selecciono una opcion valida")
```

```python
# Challenge 4
# Crear una clase animales que tenga atributo de clase y de instancia.
# Atributos de clase: especie.
# Atributos de instancia: nombre & color
# Luego, crear nuevos objetos a partir de las nuevas class. Imprimir el nombre y el color de los animales en una frase.
# Ejemplo: "Me llamo {nombre} y soy un ave de color {color}".

class Animal:
    especie = 'Animal'

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color

tucan = Animal('Tucan', 'negro')
print(f'Hola, soy un {tucan.nombre} y soy de color {tucan.color}')

tigre = Animal('Tigre', 'naranja')
print(f'Hola, soy un {tigre.nombre} y soy de color {tigre.color}')
```

```python
# Challenge 5
# Crear dos métodos para nuestro animal y ejecutar, debe ser algo específico de ese animal.
# Ejemplo: "Cuando hablo, digo X".
# Ejemplo: "Cuando me muevo, hago X".

class Mascota:
    def __init__(self, nombre, especie):
        self.nombre = nombre
        self.especie = especie


    def hablar(self, palabra):
        print(f"Soy un {self.especie} y cuando hablo digo {palabra}")

    def moverse(self, accion):
        print(f"Soy un {self.especie} y cuando me muevo, {accion}")

perro = Mascota("Puppy", "Perro")
perro.hablar("guau")
perro.moverse("corro")

pez = Mascota("Nemo", "Pez Dorado")
pez.hablar("glu glu")
pez.moverse("nado")
```

```python
# Challenge 6
# Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.
# Observación: Editar las clases hijas para que no se inicialicen.

class Vehiculo:
    def __init__(self, modelo, color):
        self.modelo = modelo
        self.color = color

class Automovil(Vehiculo):
    def presentar_automovil(self):
        print(f'El coche es un {self.modelo} de color {self.color}')    

class Motocicleta(Vehiculo): 
    def presentar_motocicleta(self):
        print(f'La moto es una {self.modelo} de color {self.color}')    

coche = Automovil('Kia Picanto', 'Rojo')
coche.presentar_automovil()

moto = Motocicleta('Kenton', 'Azul')
moto.presentar_motocicleta()
```

```python
# Challenge 7
# Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
# Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.
# Ejemplo: Yo me llamo pepito y tengo super fuerza
# Ejemplo: Yo me llamo juanito y puedo volar

class Personaje:
    def __init__(self, nombre, poder):
        self.nombre = nombre
        self.poder = poder

    def presentarse(self):
        print(f'Me llamo {self.nombre} y tengo {self.poder}')

class Heroe(Personaje):
    def salvar_la_ciudad(self):
        print(f'El superheroe {self.nombre} salvo un gato de un arbol usando su {self.poder}')


class Villano(Personaje):
    def robar_un_banco(self):
        print(f"El villano {self.nombre} robo un banco usando su {self.poder}")

superheroe1 = Heroe('Batman', 'Plata')
superheroe1.presentarse()

superheroe2 = Heroe('Flash', 'Supervelocidad')
superheroe2.salvar_la_ciudad()

villano1 = Villano('Thanos', 'Guantelete del Infinito')
villano1.presentarse()

villano2 = Villano('El Joker', 'Trucos')
villano2.robar_un_banco()
```
