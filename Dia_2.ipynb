```python
# Challenge 1
# Escribe un programa que pregunte al usuario por su edad y determine si tiene edad suficiente para comprar y consumir alcohol.
# Si el usuario tiene mas de 18 pero menos de 21 impirmir "Podes tomar pero no comprar"
# Si el usuario tiene menos de 18 imprimir "Juguito nomas"
# Si el usuario tiene mas de 21 imprimir "Hace lo que quieras"

edad = int(input("Ingrese su edad: "))

if edad >= 18:
    if edad < 21:
        print("Podes tomar pero no comprar")
    else:
        print("Hace lo que quieras")

else:
    print("Juguito nomas")
```

```python
# Challenge 2
# Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:
# Entre -10 y 18 grados, que imprima "Fresquete".
# Entre 19 y 30 grados, que imprima "Calor'i".
# Entre 31 y 45 grados, que imprima "Hakuuuuuu".
# Para todo lo dem치s, imprimir "칌amano mba'e!".
  
temperatura = int(input("Ingrese la temperatura actual: "))

if temperatura >= -10 and temperatura <= 18:
    print(f"La temperatura actual es : {temperatura} -> Fresquete")

elif temperatura >= 19 and temperatura <= 30:
    print(f"La temperatura actual es : {temperatura} -> Calor'i")

elif temperatura >= 31 and temperatura <= 45:
    print(f"La temperatura actual es : {temperatura} -> Hakuuuuuu")

else:
    print(f"La temperatura actual es : {temperatura} -> 칌amano mba'e!")
```

```python
# Challenge 7
# Solicitar al usuario que ingrese una frase.
# Imprimir un mensaje cada vez que se encuentre una letra "o"

frase = input("Ingrese una frase: ") 
  
for letra in frase:
    if letra.lower() == "o":
        print("Se encontro una letra o")
```

```python
# Challenge: WORDLE

import random

posibles_palabras = ["perro", "chipa", "limon"]
numero_de_intentos = 1
cantidad_de_intentos = 6

palabra_secreta = random.choice(posibles_palabras)

resultado = []

if numero_de_intentos <= cantidad_de_intentos:
    numero_de_intentos += 1

    palabra_ingresada = input("Ingrese una palabra: ")

    for i in range (len(palabra_ingresada)):
        if palabra_ingresada[i] == palabra_secreta[i]:
            resultado.append("九")
        elif palabra_ingresada[i] in palabra_secreta[i]:
            resultado.append("游릳")
        else:
            resultado.append("游린")

else:
    print(f"Perdiste, te quedaste sin intentos. La palabra era {palabra_secreta}")






```
