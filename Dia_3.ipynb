```python
# Challenge 2
# Crear una función que reciba dos letras (parámetros) y que verifique si son iguales:
# Si son iguales, imprimir "Las letras son iguales".
# De lo contrario, imprimir "Las letras son diferentes".

letra_1 = input("Ingrese una letra: ")
letra_2 = input("Ingrese otra letra: ")

def comparar_letras(primera_letra, segunda_letra):
    if primera_letra == segunda_letra:
        print("Las letras son iguales")
    
    else:
        print("Las letras no son iguales")

comparar_letras(letra_1, letra_2)
```

```python
# Challenge 3
# Crear una función que reciba dos parámetros (una palabra y un número), y que retorne True si el número es igual a la cantidad de letras de la palabra. 
# En caso contrario, que retorne False.

# Pista: cantidad de letras de una palabra en python

palabra = input("Ingrese una palabra: ")
numero = int(input("Ingrese un numero: "))

def comparar_cantidad_letras(palabra, numero):
    if len(palabra) == numero:
        return True
    else:
        return False
    
resultado = comparar_cantidad_letras(palabra, numero)
print(resultado)
```

```python
# Challenge 4
# Crear función que revise si una palabra contiene una letra. En caso de que la contenga, retornar la letra entre corchetes []. De lo contrario, retornar solo la letra.

palabra = input("Ingrese una palabra")
letra_ingresada = input("Ingrese una letra")

def buscar_letra(palabra, letra):
    resultado = []
    
    for letra in palabra:
        if letra == letra_ingresada:
            resultado.append(f"[{letra}]")
        else:
            resultado.append(letra)

    return resultado

busqueda = buscar_letra(palabra, letra_ingresada)
print(*(busqueda), sep="")
```

```python
# Challenge 5
# Pedir al usuario que ingrese 1 palabra, luego convertir esa palabra a una lista con sus caracteres con list(). 
# Agregar esa lista de caracteres a una lista llamada palabras_ingresadas. Repetir el proceso 3 veces. 
# Cada vez que se agregue una lista de caracteres, imprimir la lista palabras_ingresadas.
# Ejemplo del resultado esperado:
# [['c', 'i', 'e', 'l', 'o']]
#  # Al ingresar la primera palabra
# [['c', 'i', 'e', 'l', 'o'], ['c', 'a', 'l', 'o', 'r']]
#  # Al ingresar la segunda palabra
# [['c', 'i', 'e', 'l', 'o'], ['c', 'a', 'l', 'o', 'r'], ['n', 'u', 'b', 'e', 's']]
#  # Al ingresar la tercera palabra

palabras_ingresadas = []

for i in range(3):
    palabra_ingresada = input("Ingrese una palabra: ")

    palabras_ingresadas.append(list(palabra_ingresada))
    
    print(f"Cantidad de Palabras Ingresadas: {i + 1}")
    print(palabras_ingresadas)
```

```python
# Challenge 6
# Dada la lista de ejemplo: 
# lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]
# Crear una función que imprima una grilla (lista de listas), una línea por cada lista dentro de la lista principal.
# Ejemplo del resultado esperado:
# ['h', 'o', 'l', 'i', 's']
# ['h', 'e', 'l', 'l', 'o']
# ['h', 'a', 'l', 'l', 'o']

lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]

def imprimir_grilla(lista_recibida):
    for elemento in lista_recibida:
        print(elemento)

imprimir_grilla(lista)
```
